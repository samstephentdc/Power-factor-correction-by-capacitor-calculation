```python
# Power Factor improvement by calculating capacitor value

# In AC circuits there exists a difference in phase between voltage & current

# POWER FACTOR = COSINE OF PHASE DIFFERENCE

# S (APPARENET POWER) = Vol.across the element * current across the element UNIT:VA
# P (REAL POWER) = Apparent power * cos(theta) UNIT: watt
# Q (REACTIVE POWER) = Apparent power * sin(theta) UNIT:VAR(volt amp reactive)
# S^2 = P^2 + Q^2
# pf = (real power)/(apparent power)


# KVA rating of the machine is reduced by having HIGH PF.
# KVA = KW/cos(theta)


def main():
    select = input('Please select the phase value \n 1.Single phase /n 2.Three phase')
    if select == '1':
        single_phase()
    if select == '2':
        three_phase_llv()
    if select == '3':
        three_phase_lnv()
        
def single_phase():
    P_1 = float(input('Enter the real power in KW'))
    I_1 = float(input('Enter the current in amps'))
    V_1 = float(input('Enter the voltage in volts'))
    F_1 = float(input('Enter the frequency in Hertz'))
    
    # pf calculation
    pf_1 = (float(P_1)) / (float(V_1) * float(I_1))
    print('Printing Power factor:',pf_1)
    
    # apparent power calculation
    S_1 = (float(V_1) * float(I_1)) /1000
    print('Printing Apparent power:',abs(S_1))
    
    # reactive power calculation
    Q_1 = ((abs(S_1))**2 - (P_1)**2)**(1/2)
    print('Printing Reactive power:',Q_1)
    
    # power factor correction capacitance calculation
    C_1 = (1000 * Q_1)/(2*3.14*F_1*(float(V_1))**2)
    print('Printing Capacitance for Power factor calculation',C_1)
        
def three_phase_llv():
    P_2 = float(input('Enter the real power in KW'))
    I_2 = float(input('Enter the current in amps'))
    V_2 = float(input('Enter the voltage in volts'))
    F_2 = float(input('Enter the frequency in Hertz'))
    
    # pf calculation
    pf_2 = (float(P_2)) / (1.732*(float(V_2)) * float(I_2))
    print('Printing Power factor:',pf_2)
    
    # apparent power calculation
    S_2 = (1.732*(float(V_2)) * float(I_2)) /1000
    print('Printing Apparent power:',abs(S_2))
    
    # reactive power calculation
    Q_2 = ((abs(S_2))**2 - (P_2)**2)**(1/2)
    print('Printing Reactive power:',Q_2)
    
    # power factor correction capacitance calculation
    C_2 = (1000 * Q_2)/(2*3.14*F_2*(float(V_2))**2)
    print('Printing Capacitance for Power factor calculation',C_2)
    
def three_phase_lnv():
    P_3 = float(input('Enter the real power in KW'))
    I_3 = float(input('Enter the current in amps'))
    V_3 = float(input('Enter the voltage in volts'))
    F_3 = float(input('Enter the frequency in Hertz'))
    
    # pf calculation
    pf_3 = (float(P_3)) / (3*(float(V_3)) * float(I_3))
    print('Printing Power factor:',pf_3)
    
    # apparent power calculation
    S_3 = (3*(float(V_3)) * float(I_3)) /1000
    print('Printing Apparent power:',abs(S_3))
    
    # reactive power calculation
    Q_3 = ((abs(S_3))**2 - (P_3)**2)**(1/2)
    print('Printing Reactive power:',Q_3)
    
    # power factor correction capacitance calculation
    C_3 = (1000 * Q_3)/(3*2*3.14*F_3*(float(V_3))**2)
    print('Printing Capacitance for Power factor calculation',C_3)
    
main()
```
